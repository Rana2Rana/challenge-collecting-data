```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# defining lists of the target data
links_list = [] # links of one page before cleaning from noise
Address_list = []
Price_list = []
page=1 # def the counter of pages in the site 
```

```python
# Extracting the links of several pages (150 email as a sampel)
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()
for i in range(7):
   driver.get(f"https://www.immoweb.be/en/search/house-and-apartment/for-sale?countries=BE&page={page}&orderBy=relevance")
   # identify elements with tagname <a>
   lnks=driver.find_elements_by_css_selector(".card__title-link ")
   
   # traverse list
   for lnk in lnks:
      # get_attribute() to get all href
      link = lnk.get_attribute('href')
      links_list.append(link)

driver.quit()
```

```python
# saving the links in File
file_list = [links_list]
exported = itertools.zip_longest(*file_list)
with open(r"C:\Users\user\Desktop\challenge-collecting-data\links.csv","w") as myfile:
    wr = csv.writer(myfile)
    wr.writerow(['Links'])
    wr.writerows(exported)
```

```python
# finding the address
Type_property = driver.find_element_by_css_selector('.classified__title ').text
print(Type_property)
address = driver.find_element_by_css_selector(".classified__information--address-row").text
print(address)
price = driver.find_element_by_css_selector(".sr-only").text
print(price)
Bedrooms = driver.find_element_by_css_selector('.overview__text').text
print(Bedrooms)
```

```python
# creat a dictionary of the data xhich have been extracted
keys = ["Type", "Address", "Price", "Rooms"]
values = [Type_property, address, price, Bedrooms]
Data_Dict = dict(zip(keys, values))
print(str(Data_Dict))
```
